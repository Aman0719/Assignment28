```python
Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. 
    Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.
    
Ans-
```

```python
import plotly.express as px

tips = px.data.tips()

fig = px.histogram(tips, x='sex', y='total_bill', histfunc='sum',
                   color='day', pattern_shape='smoker')
fig.show()
```

```python
Q5. What is Distplot? Using Plotly express, plot a distplot.

Ans-

Distplot is a visualization tool that is used to plot a univariate distribution of observations in a dataset. 
It combines a histogram with a kernel density estimation (KDE) plot to provide a smooth estimate of the distribution of the data.
```

```python
import plotly.express as px

tips = px.data.tips()

fig = px.histogram(tips, x="total_bill", nbins=20, opacity=0.7, marginal="rug", histnorm="probability density")
fig.show()
```

```python
In the above example, we have used the tips dataset provided by Plotly Express. 
We have plotted a histogram of the 'total_bill' column using the histogram() function with the x parameter. 
The nbins parameter specifies the number of bins to use in the histogram, and the opacity parameter controls the opacity of the bars in the histogram.

The marginal parameter is set to "rug", which adds small vertical lines to the x-axis that represent each observation in the dataset. 
This allows us to see the distribution of individual observations in addition to the overall distribution.

The histnorm parameter is set to "probability density", which normalizes the histogram so that the area under the curve is equal to 1. 
This allows us to compare the shape of the distribution of the 'total_bill' column with other columns in the dataset that may have different scales or ranges.

The resulting plot should display a distplot that shows the distribution of the 'total_bill' column in the 'tips' dataset. 
The x-axis shows the range of the 'total_bill' column, and the y-axis shows the probability density of observing values in each range. 
The small vertical lines along the x-axis show the individual observations in the dataset.
```
